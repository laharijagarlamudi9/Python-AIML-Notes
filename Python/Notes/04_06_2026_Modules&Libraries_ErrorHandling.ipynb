## 04-06-2026 Modules and Libraries
As programs become larger, writing all code ourselves is difficult. Python provides pre-written code called Modules and Libraries.

* Module: A single .py file containing code.Package: A folder holding
* multiple modules and an __init__.py file.
* Library: A massive collection of related packages and modules.


### math Module

```python
# import math module
import math
```

```python
# sqrt
result = math.sqrt(64)
print(result)
```

```python
# ceil() : Rounds a number UP to the nearest integer.

print(math.ceil(4.2))
print(math.ceil(4.9))

# floor() : Rounds DOWN to the nearest integer.

print(math.floor(4.8))
print(math.floor(4.1))
```

```python
# factorial -> 5! = 5 × 4 × 3 × 2 × 1
print(math.factorial(5))

#pi -> Value of π = 3.141592653589793
radius = 5
area = math.pi * radius ** 2
print(area)
```

### DateTime Module
Used for: Current date, Current time, Date calculations, Future dates, Age calculations

```python
# import datetime module
import datetime
```

```python
# current date and time
current = datetime.datetime.now()
print(current)

# curretn date only
today = datetime.date.today()
print(today)

# current time only
current_time = datetime.datetime.now().time()
print(current_time)
```

```python
# Create Custom Date
birthday = datetime.date(2000, 8, 9)
print(birthday)
```

```python
# difference between dates

today = datetime.date.today()
birthday = datetime.date(2000, 8, 9)
difference = today - birthday
print(difference.days)
```

### Statistics
Provides statistical calculations.
Useful in: Data Analysis, Machine Learning, AI Projects

```python
# import statistics module
import statistics
```

```python
# mean - average of all values
marks = [85, 90, 78, 92, 88]
print(statistics.mean(marks))

# Median - middle value after sorting
numbers = [10, 20, 30, 40, 50]
print(statistics.median(numbers))

# Mode - Most frequently occurring value.
numbers = [10, 20, 20, 30, 40]
print(statistics.mode(numbers))

# Standard Deviation
marks = [80, 85, 90, 95, 100]
print(statistics.stdev(marks))
```

### MyMath
We can create our own modules. Suppose we repeatedly use: addition, subtraction, multiplication

Instead of writing them again and again, we create: mymath.py

```python
# Create mymath.py
# mymath.py

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b
```

```python
import mymath

print(mymath.add(10, 5))

print(mymath.subtract(10, 5))

print(mymath.multiply(10, 5))
```

```python
# Import Specific Function

from mymath import add
print(add(20, 10))
```

## Errors and Exception Handling

An error is a problem that causes a program to stop executing.

Without exception handling:

* Program crashes
* User gets ugly error message
* Remaining code does not execute

With exception handling:

* Program continues running
* Friendly message shown to user
* Application becomes more reliable

```python
# types of common errors

# --> ZeroDivisionError
print(10 / 0)

# --> ValueError, Because "abc" cannot be converted into an integer.
age = int("abc")

# --> IndexError
numbers = [10, 20, 30]
print(numbers[5])


# --> KeyError
employee = {
    "name": "Lahari"
}
print(employee["age"])
```

### try and except
The try block contains code that may produce an error.
The except block handles that error.

```python
# the ZeroDivisionError handling

try:
    num = 10

    print(num / 0)

except:
    print("Cannot divide by zero.")

print("Program Ended")

# here Program did not crash
```

```python
# User Input error
try:
    age = int(input("Enter your age: "))

    print("Age:", age)

except:
    print("Please enter numbers only.")
```

```python
# File Handling
try:
    with open("student.txt", "r") as file:
        print(file.read())

except:
    print("File does not exist.")
```

```python
# Best practice:
# Instead of: except:
# Use: except ExceptionType:
```

```python
# example
try:
    print(10 / 0)

except ZeroDivisionError:
    print("Division by zero is not allowed.")
```

```python
# real life example

try:

    balance = 5000

    amount = int(input("Enter withdrawal amount: "))

    if amount > balance:
        raise Exception("Insufficient Balance")

    print("Withdrawal Successful")

except Exception as e:
    print("Transaction Failed:", e)
```

### finally
The finally block always executes.
Whether:
Error occurs or No error occurs

Mostly used for:

* Closing files
* Closing database connections
* Releasing resources
* Logging


```python
try:
    print(10 / 0)
except:
    print("Error occurred.")
finally:
    print("Finally block executed.")
```

```python
try:

    marks = int(input("Enter marks: "))

    percentage = marks / 500 * 100

    print("Percentage:", percentage)

except ValueError:

    print("Please enter valid marks.")

finally:

    print("Thank you for using Result Calculator.")
```
