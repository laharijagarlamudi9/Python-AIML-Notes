```python
data = {
    "Student": ["Lahari", "Pranav", "Bharathi", "Bannuru", "Srikar", "Teja", "Vinay"],
    "Math": [85, 92, 78, 95, 88, 81, 94],
    "Science": [90, 89, 85, 98, 80, 87, 92],
    "English": [95, 98, 90, 99, 92, 94, 89]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.loc[1] # Returns row with label 1.
```

```python
df.iloc[1] # Returns second row.
```

```python
# Adding New Columns
df["Average"] = (df["Math"] + df["Science"]) / 2
```

```python
# Boolean Filtering : Students scoring more than 85
df[df["Math"] > 85]
```

```python
# Multiple Conditions
df[
    (df["Math"] > 80)
    &
    (df["Science"] > 85)
]
```

```python
# Selecting Specific Columns After Filtering
df[df["Math"] > 80][["Student","Math"]]
```

```python
# Reset Index : Old index becomes a new column.
df.reset_index()
```

```python
#Creates a fresh index.
df.reset_index(drop=True)
```

```python
# Set Index
df.set_index("Student")
```

```python
data = {
    "Student": ["Lahari", "Pranav", "Bharathi", "Bannuru", "Srikar", "Teja", "Vinay"],
    "Math": [85, 92, 78, 95, 88, 81, np.nan],
    "Science": [90, 89, 85, np.nan, 80, 87, 92],
    "English": [95, np.nan, 90, 99, 92, np.nan, 89]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Removes missing values, Only rows having no missing values remain.
df.dropna()
```

```python
df.dropna(axis=1) # axis=1 : Means columns.
```

```python
# thresh Parameter : Keep rows with minimum number of non-null values.

df.dropna(thresh=3)
```

```python
# fillna() : Replace missing values.
# Fill with Constant

df.fillna("Not Available")
```

```python
# Fill marks with Average marks
df["English"].fillna(
    df["English"].mean()
)
```

```python
students = pd.DataFrame({
    "Student": ["Lahari", "Pranav", "Bharathi", "Bannuru", "Srikar", "Teja", "Vinay"],
    "Department": ["Math", "Biology", "Math", "English", "Math", "Biology", "English"],
    "Total_Marks": [750, 650, 820, 680, 700, 780, 840]
})
```

```python
students.groupby("Department")
```

```python
# Average Salary by Department

students.groupby("Department").mean(
    numeric_only=True
)
```

```python
# total salary by department

students.groupby("Department").sum(
    numeric_only=True
)
```

```python
students1 = pd.DataFrame({
    "studID":[1,2,3],
    "student":["Lahari","Pranav", "Teja"]
})
```

```python
students2 = pd.DataFrame({
    "studID":[1,2,4],
    "Total_Marks":[750,650,780]
})
```

```python
pd.merge(
    students1,
    students2,
    how="inner",
    on="studID"
)
```

```python
pd.merge(
    students1,
    students2,
    how="left",
    on="studID"
)
```

```python
pd.merge(
    students1,
    students2,
    how="right",
    on="studID"
)
```

```python
pd.merge(
    students1,
    students2,
    how="outer",
    on="studID"
)
```
