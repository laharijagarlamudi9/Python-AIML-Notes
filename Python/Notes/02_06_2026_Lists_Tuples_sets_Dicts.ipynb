### LISTS
A List is a container in Python used to store multiple values in a single variable.
Ordered (maintains insertion order),
Mutable (can be modified after creation),
Allows duplicate values,
Can store different data types,
Uses square brackets [],
Supports indexing and slicing.

```python
Names = ["Govinda", "Lahari", "Jagarlamudi", "08/09", 10, 2.99]
print(Names)
```

### INDEXING AND SLICING

Indexing is used to access a single element from a list.

Slicing is used to access multiple elements from a list.

```python
fruits = ["apple", "banana", "orange", "grapes"]

print(fruits[0])
print(fruits[1])
print(fruits[3])
```

```python
fruits = ["apple", "banana", "orange", "grapes"]

print(fruits[-1]) # last element
print(fruits[-2]) # second last element
```

```python
# slicing
# list_name[start:end], Start included, End excluded

numbers = [10,20,30,40,50,60]

print(numbers[1:4])
print(numbers[2:])
print(numbers[::-1])

# Index starts from 0
# Negative index starts from -1
# Slicing does not modify original list
# End index is never included
```

### CONCATENATING LISTS
Combining two or more lists into a single list.

```python
list1 = [1,2,3]
list2 = [4,5,6]

result = list1 + list2

print(result)

# Original lists remain unchanged.
# Creates a new list.
```

### DOUBLE THE LIST USING *
The * operator repeats list elements multiple times.

```python
Names = ["Lahari", "Pranav"]

print(Names * 3)

# Does not multiply values.
# Repeats the entire list.
```

## BASIC LIST METHODS
### append() : Adds one element at the end of the list.

```python
L = [1,4,7,3,8,9]
L.append(10)
L
```

### pop() : Removes and returns an element.
by default, pop removes the lastly added element from the list, unless specified.

```python
numbers = [10,20,30]

removed = numbers.pop()

print(removed)
print(numbers)
```

```python
numbers = [10,20,30]

removed = numbers.pop(1)

print(removed)
print(numbers)
```

### sort() : Sorts the list in ascending order.

```python
numbers = [15,1,23,32,7]

numbers.sort()

print(numbers)
```

### reverse() : Reverses the list permanently.

```python
numbers = [15,1,23,32,7]
numbers.reverse()

print(numbers)
```

## NESTED LISTS

A list inside another list is called a Nested List.

```python
details = [["Lahari", 25], ["Pranav", 24], ["Pandu", 21]]
print(details)
print(details[0])
print(details[1][1])


# First index selects inner list.
# Second index selects element inside inner list.
```

## LIST COMPREHENSIONS
A shorter way to create lists using loops.

```python
# this is the normal way we usually do
numbers = []

for num in range(1,6):
    numbers.append(num)

print(numbers)
```

```python
# in comprehension way

numbers = [num for num in range(1,6)]

print(numbers)
```

## ADVANCED LIST METHODS

```python
# append()
numbers = [1,2,3]
numbers2 = [4,5]

numbers.append(numbers2)

print(numbers)
```

```python
# count() : Returns number of occurrences.

numbers = [1,2,2,3,2]

print(numbers.count(2))
```

```python
# remove() : Removes by value.
numbers = [10,20,30,40]

numbers.remove(20)

print(numbers)
```

## TUPLES
A Tuple is a collection used to store multiple values in a single variable. A tuple is very similar to a list, but the major difference is: Tuples are immutable (cannot be modified after creation).

Characteristics of Tuples :
Ordered,
Immutable,
Allows duplicates,
Supports indexing and slicing,
Faster than lists,
Written using parentheses ()

```python
# basic syntax
# tuple_name = (value1, value2, value3)
```

```python
# creating a tuple
names = ("Lahari", "Bharathi", "Sruthi", "Pranav", "Pandu")
print(names)
print(names[0])
print(names[1])
print(names[-1])
```

```python
# Slicing
names = ("Lahari", "Bharathi", "Sruthi", "Pranav", "Pandu")
print(names[1:4])
```

```python
# tuple packing
employee = ("Lahari", 25, "IT")
```

```python
# tuple unpacking : Extracting values from a tuple.
employee = ("Lahari", 25, "IT")

name, age, dept = employee

print(name)
print(age)
print(dept)
```

```python
# we cannot modify tuples
numbers = (10,20,30)

numbers[0] = 100
```

```python
# single value tuple

num1 = (19) # -> this is wrong, because this will be taken as integer.
num2 = (19,) # -> this is the right way
```

### Useful Tuple Methods

```python
# count()
numbers = (1,2,2,3,2)

print(numbers.count(2))
```

```python
# index()
numbers = (10,20,30)

print(numbers.index(20))
```

## SETS
A Set is an unordered collection of unique values.
Sets automatically remove duplicate values.

Characteristics of Sets, Unordered, Mutable, No duplicates allowed, No indexing, Written using curly braces {}

```python
# Syntax : set_name = {value1, value2, value3}
```

```python
# creating a set
numbers = {10,20,30}

print(numbers)
```

```python
# Duplicate Removal
numbers = {10,20,20,30,30}

print(numbers)
```

```python
#Adding Elements

numbers = {10,20}

numbers.add(30)

print(numbers)
```

```python
# removing elements

numbers = {10,20,30}

numbers.remove(20)

print(numbers)
```

### SET Operations

```python
# Union : Combines all unique values.
A = {1,2,3}
B = {3,4,5}

print(A | B)
```

```python
# Intersection : Common values only.

A = {1,2,3}
B = {2,3,4}

print(A & B)
```

```python
# Difference : Values present in first set but not second.

A = {1,2,3}
B = {2,3,4}

print(A - B)
```

```python
# Symmetric Difference : Values not common in both.

A = {1,2,3}
B = {2,3,4}

print(A ^ B)
```

```python
# add()
numbers = {1,2}

numbers.add(3)
print(numbers)
```

```python
# update() : Add multiple values.

numbers = {1,2}

numbers.update([3,4,5])
print(numbers)
```

```python
# remove()
numbers.remove(2)
print(numbers)

#Error if value not found.
```

```python
# discard()

numbers.discard(100)
print(numbers)

# even though there is no 100 in the set, it will not return error.
```

```python
# pop() : Removes a random element.

numbers.pop()
print(numbers)
```

```python
# clear() : Removes everything.
numbers.clear()
print(numbers)

# this returns an empty set
```

## DICTIONARIES
A Dictionary stores data in the form of Key : Value

Characteristics of Dictionaries: Mutable, Ordered (Python 3.7+), Keys must be unique, Values can repeat, Written using {}

```python
# syntax
#dict_name = {
#    key1:value1,
#    key2:value2
#}
```

```python
employee = {
    "name":"Lahari",
    "age":25,
    "Dept":"IT"
}
print(employee)  # -> gives the complete dict
print(employee["name"]) # -> trying to accessing a value
print(employee.get("age")) # -> Using get()
```

```python
# adding new data
employee = {
    "name":"Pranav"
}

employee["age"] = 24

print(employee)
```

```python
# Updating Existing Data
employee = {
    "age":25
}

employee["age"] = 30

print(employee)
```

```python
# Removing Data : pop()
employee = {
    "name":"Lahari",
    "age":25,
    "Dept":"IT"
}

employee.pop("age")

print(employee)
```

### Dictionary Methods

```python
# keys() and values()
employee = {
    "name":"Lahari",
    "age":25,
    "Dept":"IT"
}

print(employee.keys())
print(employee.values())
```

```python
# items()

print(employee.items())
```

```python
# Looping Through Dictionary
# Keys
employee = {
    "name":"Lahari",
    "age":25,
    "Dept":"IT"
}

for key in employee:
    print(key)

# Values
for value in employee.values():
    print(value)

# Key and Value Together
for key, value in employee.items():
    print(key, value)
```
