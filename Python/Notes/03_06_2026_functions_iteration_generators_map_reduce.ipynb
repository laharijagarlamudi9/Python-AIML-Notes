### Return Statement
a return statement exits a function and sends a specific value back to the place where that function was originally called. It marks the official end of the function's execution, meaning any lines of code placed directly after it inside the block are ignored.

In languages like Python or JavaScript, if you do not explicitly include a return keyword, the function automatically outputs an empty value like None or undefined.

### Difference Between Return and Print:

print simply displays text on the screen for a human to read. The computer cannot actually use or reuse that data elsewhere in the program logic.

return invisibly passes the data directly to the computer's memory so your program can continue working with it.

```python
# without return

def add(a, b):
    print(a + b)

result = add(10, 20)

print(result)

# Function printed 30. But it returned nothing. So Python stores None.
```

```python
# using return

def add(a, b):
    return a + b

result = add(10, 20)

print(result)

# returned values can be reused.
```

```python
# returning multiple values

def employee():
    return "Lahari", 25, "IT"

name, age, dept = employee()

print(name)
print(age)
print(dept)
```

```python
# returning boolean

def is_even(num):

    if num % 2 == 0:
        return True

    return False

print(is_even(10))
```

### ITERATORS
An iterator in Python is an object that implements the iterator protocol, allowing you to traverse a collection one element at a time. It consists of two essential dunder methods: __iter__(), which returns the iterator object itself, and __next__(), which fetches the next available value. When no items remain, the iterator raises a StopIteration exception to safely terminate the loop.

```python
# Python internally moves through the list one item at a time.

nums = [10, 20, 30]

it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
```

```python
# what will happen after last element ?
print(next(it))

# it throws error with 'stop Itereation' because there are no more values.
```

```python
names = ["Lahari", "Pranav", "Bharathi"]

# 1. Get the iterator object from the container
name_iterator = iter(names)

# 2. Loop indefinitely to fetch items
while True:
    try:
        # 3. Get the next item
        message = next(name_iterator)
        print(f" Welcome {message}, Happy to see you !")
    except StopIteration:
        # 4. Stop the loop cleanly when elements run out
        break
```

### GENERATORS
Python generators are special functions that return a lazy iterator, allowing you to stream or loop through data one item at a time without loading the entire dataset into memory. Instead of using return to terminate and send back a single value, generators use the yield keyword to temporarily pause execution, remember the function's exact state, and pass a value back to the caller.

Instead of return, it uses yield.

Normal functions store everything in memory. Generators generate values only when needed. Memory efficient.

```python
# Entire list created at once.
def numbers():
    return [1, 2, 3]
```

```python
def numbers():

    yield 1
    yield 2
    yield 3

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def count():

    for i in range(1, 6):
        yield i

num = count()

for n in num:
    print(n)
```

### MAP
Applies a specified function to each item of an iterable (like a list, tuple, or dictionary) and returns a lazy generator object. Because it returns an iterator, you must explicitly cast it to a sequence type like list() or tuple() to view or work with the resulting data directly.


syntax : map(function, iterable)

The first argument is the name of a function and the second a sequence (e.g. a list). map() applies the function to all the elements of the sequence. It returns a new list with the elements changed by the function.

```python
# You can use map() with any built-in or user-defined function.
names = ["Lahari", "Pranav", "Bharathi"]

# Apply the str.upper method to each item
uppercase_names = map(str.upper, names)

# Convert the iterator back to a list
print(list(uppercase_names))
```

```python
# Using map() with a Lambda Function

numbers = [1, 2, 3, 4, 5]

# Square each number in the list
squared = map(lambda x: x ** 2, numbers)

print(list(squared))
```

```python
# Adding 18% GST to product prices.

prices = [100, 200, 300]

gst_prices = map(
    lambda p: p * 1.18,
    prices
)

print(list(gst_prices))
```

### REDUCE()

reduce() function (from the functools module) applies a function cumulatively to the elements of an iterable and returns a single final value. It processes elements step-by-step, combining two elements at a time until only one result remains.

```python
# syntax
from functools import reduce

reduce(function, iterable)

# function: A binary function taking exactly two arguments.
# iterable: A collection of elements (e.g., list, tuple).
# initializer: An optional starting value placed before the first element.
```

```python
from functools import reduce

nums = [1,2,3,4]
result = reduce(
    lambda x, y: x + y,
    nums
)
print(result)
```

When executing reduce(lambda x, y: x + y, [1, 2, 3, 4]):It takes the first two elements: x = 1, y = 2. It returns 1 + 2 = 3.It uses the result as the next x and takes the third element: x = 3, y = 3. It returns 3 + 3 = 6.It uses that result as x and takes the final element: x = 6, y = 4. It returns 6 + 4 = 10.The list is empty, so it outputs the single final value: 10.

```python
# Find largest number.

from functools import reduce

nums = [10, 50, 20, 70, 30]

largest = reduce(
    lambda a, b: a if a > b else b,
    nums
)

print(largest)
```

### FILTER()
The filter() function in Python extracted elements from an iterable sequence (such as a list, tuple, or set) that satisfied a specific condition. It evaluates each item against a function and yields a lazy iterator object containing only the items that evaluated to True.

The standard syntax : filter(function, iterable)

*   function: A function that tests each element. It returns True to keep the item or False to discard it.
*   iterable: The collection of items to be filtered.



```python
# Keep only even numbers.

nums = [1,2,3,4,5,6]

result = filter(
    lambda x: x % 2 == 0,
    nums
)

print(list(result))
```

```python
numbers = [10, 15, 20, 25, 30]
greater_than_20 = list(filter(lambda x: x > 20, numbers))
print(greater_than_20)
```
