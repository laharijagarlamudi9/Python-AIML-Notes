## 04-06-2026 FILE OPERATIONS

File operations mean working with files using Python. We can create files, write data into files, read data from files, append new data, copy files, and modify file content. In Python, we mainly use the built-in open() function.

file = open("file_path", "mode")

Common modes:

* "w" : write mode - creates a file or overwrites existing content
* "r" : read mode - reads an existing file
* "a" : append mode - adds content at the end
* "r+": read and write mode

```python
#A better and safer way is using with open() because it automatically closes the file.

# with open("file_path", "mode") as file:
#      file operations here
```

### Writing to a text file
Writing means creating a file and adding content into it. We use "w" mode for writing.

"w" mode overwrites the existing content. That means if the file already has some data, it will be deleted and replaced with the new data.

```python
# File path where we want to create the text file
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

# Opening the file in write mode
# If the file does not exist, Python will create it
# If the file already exists, Python will overwrite it
with open(file_path, "w") as file:
    file.write("Hello Lahari!\n")
    file.write("Welcome to Python file operations.\n")
    file.write("Today we are learning how to write into a text file.\n")

print("File written successfully.")
```

### Reading from a text file
Reading means opening an existing file and reading its content. We use "r" mode.

Note: If the file does not exist, Python gives an error. 'FileNotFoundError'

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

# Opening the file in read mode
with open(file_path, "r") as file:
    content = file.read()

print(content)
```

### Reading Line-by-Line Using For Loop

Sometimes we do not want to read the complete file at once.
Instead, we can read one line at a time using a for loop.
This is useful when the file has many lines.

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    # The for loop reads the file line by line
    for line in file:
        print(line)

# You may see extra blank lines because every line already has \n, and print() also adds a new line.
```

```python
# To remove extra space:
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    for line in file:
        print(line.strip())
```

### Append

Append means adding new content at the end of an existing file. We use "a" mode. Append mode does not delete old content.

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

# Opening file in append mode
with open(file_path, "a") as file:
    file.write("This line is added using append mode.\n")
    file.write("Append mode adds content at the end.\n")

print("New content appended successfully.")
```

### First n Characters

Sometimes we do not want to read the full file. We may want to read only a few characters.
For that, we use: "file.read(number_of_characters)"

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    first_12_chars = file.read(12)

print("first 12 chars: ", first_12_chars)

# Python counts spaces also as characters.
```

### Reading Line by Line Using .readline()

.readline() reads only one line at a time.
Every time we call .readline(), Python moves to the next line.
.strip() removes extra spaces and newline characters from the beginning and end of a string.

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    line1 = file.readline()
    line2 = file.readline()
    line3 = file.readline()

print(line1)
print(line2)
print(line3)
```

```python
# Now using .strip():

file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    line1 = file.readline().strip()
    line2 = file.readline().strip()
    line3 = file.readline().strip()

print(line1)
print(line2)
print(line3)
```

### Read Multiple Lines and Put That Into a List
To read all lines from a file and store them in a list, we use: file.readlines(). Each line becomes one item in the list.

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r") as file:
    lines_list = file.readlines()

print(lines_list)
```

### Copy File
Copying a file means reading content from one file and writing that same content into another file.

We need two files:
* source file      # original file
* destination file # copied file

```python
source_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"
destination_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/copy_sample.txt"

# Open source file in read mode
with open(source_path, "r") as source_file:
    content = source_file.read()

# Open destination file in write mode
with open(destination_path, "w") as destination_file:
    destination_file.write(content)

print("File copied successfully.")
```

### r+ Mode
r+ mode means reading and writing together.

Important points:
* r+ mode needs the file to already exist.
* It does not create a new file.
* It starts writing from the beginning of the file.
* It can overwrite existing characters.

```python
file_path = "/Users/jagarlamudilahari/Desktop/Python_AIML/Python/sample.txt"

with open(file_path, "r+") as file:
    # First read the existing content
    content = file.read()
    print("Before writing:")
    print(content)

    # Move cursor to the beginning of the file
    file.seek(0)

    # Write new content from the beginning
    file.write("Hi Lahari!\n")

print("r+ operation completed.")
```

### Writing Data to an Excel File

We first create some data and convert it into a DataFrame and then save it as an Excel file.

```python
# Import pandas library
import pandas as pd

# Sample student data
data = {
    "Name": ["Alice", "Bob", "Charlie", "Ruchik", "Sushma", "Thauja", "Max", "Karthik"],
    "Marks": [85, 90, 78, 87, 11, 44, 0, 21]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print("Original Data:")
print(df)

# Save DataFrame to Excel file
# index=False prevents pandas from creating an extra index column
df.to_excel("students.xlsx", index=False)

print("\nExcel file created successfully!")
```
