## 08-06-2026 DATA VISUALIZATION
Imagine you have thousands of rows of data in an Excel sheet.

Looking at numbers alone makes it difficult to: Identify trends, Compare values, Find patterns, Detect outliers, Make decisions quickly.

You can read the data, but understanding growth takes effort. With Visualization a graph instantly shows the trends.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Student": ["Lahari", "Pranav", "Bharathi", "Bannuru", "Srikar", "Teja", "Vinay"],
    "Math": [85, 92, 78, 95, 88, 81, 94],
    "Science": [90, 89, 85, 98, 80, 87, 92],
    "English": [95, 98, 90, 99, 92, 94, 89]
}

df = pd.DataFrame(data)

print(df)
```

```python
# head - Used to display the first few rows if specified. by default it displays first 5 rows
df.head()
```

```python
print(df.head(3))
```

```python
# type() - Used to check the datatype of an object.
print(type(df))
```

```python
print(type(df["Math"]))
```

```python
# plot() - Used to create graphs quickly.

#Plot Math Scores Only

df["Math"].plot()
plt.show()
```

```python
# plot with labels - Labels make graphs readable.

df.plot(
    x="Student",
    y="Math",
)

plt.title("Math Scores of Students")
plt.xlabel("Student Index")
plt.ylabel("Math Marks")

plt.show()
```

```python
# Bar plots are useful for comparisons.

df.plot(
    x="Student",
    y="Math",
    kind="bar"
)

plt.title("Math Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
# Multiple Columns in Bar Plot
df.plot(
    x="Student",
    y=["Math", "Science"],
    kind="bar"
)

plt.show()

#Each student gets: One bar for Math, One bar for Science
```

### Using Built-in Dataset - Tips Dataset from Seaborn.

```python
tips = sns.load_dataset("tips")

print(tips.head(100))
```

### understanding of data set

| Column     | Meaning          |
| ---------- | ---------------- |
| total_bill | Restaurant bill  |
| tip        | Tip given        |
| sex        | Male/Female      |
| smoker     | Yes/No           |
| day        | Day of visit     |
| time       | Lunch/Dinner     |
| size       | Number of people |


```python
# Plot Total Bill
tips["total_bill"].plot()

plt.title("Restaurant Bills")
plt.ylabel("Bill Amount")

plt.show()
```

```python
# Average Bill by Day
tips.groupby("day")["total_bill"].mean().plot(
    kind="bar"
)

plt.title("Average Bill per Day")
plt.xlabel("Day")
plt.ylabel("Average Bill")

plt.show()
```

```python
# Sometimes a column is not useful.We can remove it.
#Syntax   :     df.drop("column_name", axis=1)

new_df = tips.drop("sex", axis=1) # axis = 1 Means remove a column.
print(new_df.head())
```

```python
print(tips.columns)
```

```python
# iloc means: Integer Location
# Used to access rows and columns using index numbers.
# Syntax   :  df.iloc[row, column]


print("# Get First Row")
print(tips.iloc[0])

print("# Get First Row, Second Column")
print(tips.iloc[0, 1])

print("# first 5 rows")
print(tips.iloc[0:5])

print("#First 5 Rows and First 3 Columns")
print(tips.iloc[0:5, 0:3])
```

```python
# iloc + Plot : Plot only selected columns.
# Plot Total Bill and Tip using iloc

# Select only numeric columns:
#total_bill → position 0, tip → position 1

tips.iloc[:, 0:2].plot()

plt.title("Total Bill and Tip")
plt.xlabel("Customer Index")
plt.ylabel("Amount ($)")
plt.show()
```

```python
# Bar Plot using iloc
# Let's compare the first 10 customers' bills.

tips.iloc[0:10, [0]].plot(
    kind="bar",
    legend=False
)

plt.title("Total Bill - First 10 Customers")
plt.xlabel("Customer Index")
plt.ylabel("Bill Amount")
plt.show()
```

```python
# Multiple Columns Bar Plot : Compare Total Bill and Tip together.

tips.iloc[0:10, 0:2].plot(
    kind="bar"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Customer Index")
plt.ylabel("Amount ($)")
plt.show()


# Each customer gets: One bar for Total Bill, One bar for Tip
```

```python
df = pd.DataFrame(
    np.random.randint(10, 100, size=(10, 4)),
    columns=['Product_A', 'Product_B', 'Product_C', 'Product_D']
)

print(df.head())
```

```python
df.plot.bar()
```

```python
df.plot.barh()
```

```python
df.plot.hist()
```

```python
# Bins are Number of intervals, Alpha controls transparency.
# 1.0 → Fully visible, 0.5 → Semi transparent, 0.2 → More transparent

df.hist(
    bins=10,
    color='yellow',
    alpha=0.5
)

plt.show()
```

```python
# DIFF() : Calculates difference between current and previous row.
print(df.diff())
```

```python
df.diff().plot()

plt.title("Difference Plot")
plt.show()
```

```python
# Scatter plots compare two columns.

tips.plot.scatter(
    x='total_bill',
    y='tip',
    cmap='viridis'
)
plt.show()
```

```python
# Hexagon plot : Useful for large datasets. Groups nearby points into hexagons.
tips.plot.hexbin(
    x='total_bill',
    y='tip',
    gridsize=10,
    cmap='coolwarm'
)
plt.show()
```

```python
# Pie charts need one row or one series.
df.iloc[0].plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

plt.ylabel("")
plt.show()
```

```python
#autopct='%1.1f%%' : Show percentages
#startangle=90 : Rotate chart
#shadow=True : Add shadow
#explode=[0.1,0,0,0] : Separate slice

df.iloc[0].plot(
    kind='pie',
    autopct='%1.1f%%',
    explode=[0.1,0,0,0]
)

plt.show()
```

```python
# SCATTER MATRIX : Compare every column against every other column.

from pandas.plotting import scatter_matrix

scatter_matrix(
    tips,
    figsize=(10,10)
)

plt.show()
```

```python
# Scatter Matrix with Transparency
scatter_matrix(
    tips,
    figsize=(10,10),
    alpha=0.8
)

plt.show()
```

```python
# Line Plot with Markers
df.plot(
    marker='o'
)

plt.show()
```

```python
# Curved Style
df.plot(
    linestyle='--',
    marker='o'
)

plt.show()
```

```python
# SUBPLOTS : Display each column separately.

df.plot(
    subplots=True,
    figsize=(10,8)
)

plt.show()
```

```python
# subplots with layout
df.plot(
    subplots=True,
    layout=(2,2),
    figsize=(10,6)
)

plt.show()
```
